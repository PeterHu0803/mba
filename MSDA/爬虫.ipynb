## 链接的类型

链接一般分为两大类：GET和POST

- GET：仅提供链接地址url即可，如果包含参数则拼接在url中
- POST：除了链接地址url之外，如果有参数则包含在请求体（body）中，而不是直接拼接在url中，即从url中看不到传递的参数，因此更为安全

````python
url = 'http://maoyan.com/board/4'
response = requests.get(url)

# 这是一个requests的response
print('=' * 30)
print(type(response))

# response请求状态码，200表示成功，一些其他常见的状态码：400、403、404、502等
print('=' * 30)
print(response.status_code)

# response的内容，是一个字符串
print('=' * 30)
print('Response Content')
print(response.text)

# response支持cookie，cookie即客户端缓存，模拟登陆时会用到，用于保存登陆的账号、密码、登陆状态等信息
print('=' * 30)
print('Response Cookie')
print(response.cookies)
````

## 在GET链接中传入参数

在访问GET类型链接时，可以根据网站需要传入一些参数，可以传一个或多个参数，遵循以下格式，`url?key1=value1&key2=value2`

例如[http://maoyan.com/board/4](http://maoyan.com/board/4)是Top100的第一页，[http://maoyan.com/board/4?offset=90](http://maoyan.com/board/4?offset=90)是Top100的最后一页

即offset这个参数表示 **偏移量**，或者说当前页10部电影的起始编号

以下代码获取猫眼电影Top100榜的全部电影信息，爬取完毕后写入文件

````python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36',
    'Cookie' : '__mta=150386953.1599996038849.1600045310969.1600053979767.11; uuid_n_v=v1; uuid=24972DE0F5B311EAAED6A93BD058FBB6218CB2A77AA84DD8B8FD0B65E320307E; _csrf=6171e7b7d7f5a2f2597de153ab2f3b89118ec979255a5fbf47f0321085688c00; mojo-uuid=35c5d332b75828b5d6d74a998beb230b; _lxsdk_cuid=17487320bdcc8-011ee530122375-316b7004-1aeaa0-17487320bdcc8; _lxsdk=24972DE0F5B311EAAED6A93BD058FBB6218CB2A77AA84DD8B8FD0B65E320307E; Hm_lvt_703e94591e87be68cc8da0da7cbd0be2=1599996038,1600002186; Hm_lpvt_703e94591e87be68cc8da0da7cbd0be2=1600053979; mojo-session-id={"id":"ca7e2715497ecd402697927b3c69488d","time":1600053979406}; mojo-trace-id=1; _lxsdk_s=1748aa62905-c5d-e74-c0e%7C%7C2'
}

with open ('./data/maoyan_top100.txt', 'w') as fw:
    
    for i in range(10):
        # 根据offset调整链接
        url = 'http://maoyan.com/board/4?offset=' + str(i * 10)
        response = requests.get(url, headers = headers)
        print(url)
        text = response.text
        html = BeautifulSoup(response.text, 'html.parser')
    
        titles = html.find_all('p', attrs = {'class': 'name'})
        titles = [i.get_text() for i in titles]
    
        stars = html.find_all('p', attrs = {'class': 'star'})
        stars = [i.get_text().strip().lstrip('主演：') for i in stars]
    
        dates = html.find_all('p', attrs = {'class': 'releasetime'})
        dates = [i.get_text().lstrip('上映时间：') for i in dates]
    
        scores = html.find_all('p', attrs = {'class': 'score'})
        scores = [i.get_text() for i in scores]
    
        imgs = html.find_all('img', attrs = {'class': 'board-img'})
        imgs = [i.get('data-src') for i in imgs]
    
        for i in range(len(titles)):
            print(titles[i])
            fw.write('\t'.join([titles[i], stars[i], dates[i], scores[i], imgs[i]]) + '\n')
````

恭喜，你已经掌握了用Python爬取数据的基本能力

但别高兴得太早，以上代码能成功运行的前提是，网站不设置任何防爬措施

实际上，为了获取大量数据，例如100W部电影的基本信息，我们可能需要用爬虫频繁请求某个链接

在网站服务器看来，同一个ip如此频繁地请求，肯定不是正常用户，封ip！

解决封ip的简单方法是降低爬取频率，例如每处理完一个链接后，使用`time.sleep(1)`等待1秒，花更多的时间，避免被封

比较高级的方法是使用代理或ip池，每次被封ip后，就换一个新的ip继续爬，看上去就好像是很多正常用户在访问一样

但道高一尺、魔高一丈，没有无敌的爬虫，也没有无敌的防爬，无非是看哪一边花的时间和精力更多
