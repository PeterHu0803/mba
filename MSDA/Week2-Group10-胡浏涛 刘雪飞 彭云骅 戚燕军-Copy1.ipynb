```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pandas.core.frame import DataFrame
url = 'https://www.sjtu.edu.cn/yxsz/'
def get_college_info(link_url):
    response = requests.get(link_url)
    soup = BeautifulSoup(response.text,'html.parser') #解析数据
    links_div = soup.find_all('li',class_ = "col-sm-4")+soup.find_all('li',class_ = "col-md-6")+soup.find_all('li',class_ = "col-sm-6")#整理数据
    print(links_div)
    names=[name.a.text for name in links_div] #获取学院名称
    links=[link.a.get('href') for link in links_div] #获取学院网址
    college_info={"学院名称" : names, "学院网址" : links}#将列表names,links转换成字典
    college_info_data=DataFrame(college_info)#将字典转换成为数据框
    return college_info_data
get_college_info(url)
college_info_data=get_college_info(url)
college_info_data.to_csv('上海交通大学院系网址.csv',encoding='utf_8_sig',index=None)
```
